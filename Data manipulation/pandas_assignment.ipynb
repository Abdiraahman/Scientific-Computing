Basic Statistics

```python
#mean
data1["Final_Score"].mean()
```

```python
#Median
data1["Final_Score"].median()
```

```python
#Mode
data1["Final_Score"].mode()
```

```python
#standard deviation
data1["Final_Score"].std()
```

```python
#first quartile
Q1 = data1["Final_Score"].quantile(0.25)
Q1
```

```python
#Third quartile
Q3 = data1["Final_Score"].quantile(0.75)
Q3
```

```python
#Inter quartile range
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1-1.5*IQR
```

```python
upper_bound = Q3 + 1.5*IQR
```

```python
#checking outliers
data_no_outlier=data1[~((data1['Final_Score']<Q1-1.5*IQR))|data1['Final_Score']>Q3+1.5*IQR]
data_no_outlier
```

```python
# scatterplot
sns.scatterplot(data=data1, x="Assignments_Avg", y="Final_Score", color="black")

plt.title("Scatter Plot: Assignments_Avg vs Final_Score")  
plt.show()
```

Correlation between data

```python
#correlation between age and final score
data1Num = data1[["Age","Final_Score"]]
data1Num.corr()
```

```python
numeric_data = data1.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()
correlation_matrix
```

```python
# correlation heatmap 
plt.figure(figsize=(10, 7))  
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
